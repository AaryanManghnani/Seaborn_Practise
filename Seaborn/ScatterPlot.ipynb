```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('penguins')
df
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=df)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=df,
                hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=df,
                hue="sex",
                style="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=df,
                hue="sex",
                style="sex",
                size="sex",
                sizes=(20,30))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=df,
                hue="sex",
                style="sex",
                size="sex",
                sizes=(20,30),
                palette="gist_ncar")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=df,
                hue="sex",
                style="sex",
                size="sex",
                sizes=(20,30),
                palette="gist_ncar")
plt.show()
```
