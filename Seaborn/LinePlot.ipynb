Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
df = pd.DataFrame({"var":var, "var_1":var_1})
df
```

```python
sns.lineplot(x="var",y="var_1", data=df)
plt.show()
```

```python
# or you can do
sns.lineplot(x=var,y=var_1)
plt.show()
```

```python
print(sns.get_dataset_names())
```

```python
df1 = sns.load_dataset('penguins')
df1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data=df1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data=df1, hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data=df1, hue="sex",size=10)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data=df1, hue="sex", style="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data=df1, hue="sex", style="sex", palette="Accent")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data=df1, hue="sex", style="sex", palette="Accent", markers=["o",">"])
plt.show()
```

```python
df1 = df1.head(20)
df1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data=df1, hue="sex", style="sex", palette="Accent", markers=["o",">"], dashes=False, legend=False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm", data=df1, hue="sex", style="sex", palette="Accent", markers=["o",">"], dashes=False, legend=False)
plt.grid()
plt.title("Graph")
plt.show()
```
